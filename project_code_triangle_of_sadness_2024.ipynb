```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler=StandardScaler()
Xn_train=scaler.fit_transform(X_train)
Xn_test=scaler.transform(X_test)
```

# Model 1 : Logistic Regression

```python
model1=LogisticRegression()
model1.fit(Xn_train,y_train)

y_pred1=model1.predict(Xn_test)

print(classification_report(y_test, y_pred1))
```

**We choose recall as the most accurate metric for this problem 
as we want to detect as much as possible positive values.**

```python
recall1=recall_score(y_test, y_pred1)
print(f"Recall Score: {recall1}")
```

```python
print("Confusion Matrix (Test):")
matrix1=confusion_matrix(y_test, y_pred1)
print(matrix1)
matrix1 =ConfusionMatrixDisplay(confusion_matrix=matrix1)
matrix1.plot(cmap="viridis")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

# Grid Search


```python
param_grid1 = [ 
    {   'classifier': [SVC()],
        'classifier__C': [0.001, 1, 100, 500, 1000],
        'classifier__kernel': ['linear', 'poly', 'rbf', 'sigmoid']},
    
    {   'classifier': [LogisticRegression()],
        'classifier__C': [0.001, 1, 100, 500, 1000],
        'classifier__penalty': ['l1', 'l2'],
        'classifier__solver': ['liblinear']},  # liblinear υποστηρίζει μόνο l1 και l2 penalty
    
    {   'classifier': [LogisticRegression()],
        'classifier__C': [0.001, 1, 100, 500, 1000],
        'classifier__penalty': ['l2', 'none'],
        'classifier__solver': ['lbfgs']},  # lbfgs υποστηρίζει μόνο l2 και none penalty
    
    {   'classifier': [LogisticRegression()],
        'classifier__C': [0.001, 1, 100, 500, 1000],
        'classifier__penalty': ['l1', 'l2'],
        'classifier__solver': ['saga']},  # saga υποστηρίζει l1, l2, elasticnet και none penalty
]

print('\n')

pipeline1 = Pipeline([  
    ('classifier', SVC())  ])
    
grid_search = GridSearchCV(pipeline1, param_grid1, cv=8, n_jobs=-1, scoring='recall', verbose=4)

grid_search.fit(Xn_train, y_train)
print("Best Parameters:", grid_search.best_params_)
print('\n')
print("Best Model Recall:", grid_search.best_score_)


y_pred_grid = grid_search.best_estimator_.predict(Xn_test)
recall_test_grid = recall_score(y_test, y_pred_grid)
print("Recall on Test Set (GridSearch Best Model):", recall_test_grid)
```

**Confusion Matrix (GridSearch)**
